```python
day_df = pd.read_csv('station_data.csv')
```

```python
day_df = day_df.drop(["Unnamed: 0", "DAILY_ENTRIES", "DAILY_EXITS"], axis=1, errors="ignore")
```

```python
#combine Station and linename
day_df['STATION_LINE'] = day_df['STATION'] + ' ' + day_df['LINENAME']
```

```python
day_df = day_df.drop(["STATION", "LINENAME"], axis=1, errors="ignore")
```

```python
day_df.to_csv('station_data.csv')
```

```python
day_df
```

```python
(day_df
 .groupby(['DAYS', 'STATION_LINE'])
 .sum()
 .reset_index()
 .sort_values(['TOTAL_TRAFFIC'], ascending=[False])
)
```

```python
(day_df
 .groupby(['DAYS'])
 .sum()
 .reset_index()
 .sort_values(['TOTAL_TRAFFIC'], ascending=[False])
)



```

```python
day_df2 = day_df.groupby(['DAYS', 'STATION_LINE'])[['TOTAL_TRAFFIC']].sum().reset_index().sort_values(['TOTAL_TRAFFIC'], ascending=False)
```
